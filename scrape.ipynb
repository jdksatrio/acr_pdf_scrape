## EXTRACT UNIQQUEID FROM THE HTML ##

```python
from bs4 import BeautifulSoup

html_file = "./prereq/Appropriateness Criteria.html"

with open(html_file, "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

unique_docid = []

for a in soup.find_all("a"):
    text = a.get_text(strip=True)
    if "Narrative &" in text:
        href = a.get("href")
        docid = href.split("/")[2]
        if docid not in unique_docid:
            unique_docid.append(docid)
```

## DOWNLOAD THE PDFS FROM KNOWN IDS ##

```python
import requests
from bs4 import BeautifulSoup
import os
import time

os.makedirs("acr_pdfs", exist_ok=True)


known_ids = unique_docid

def download_pdf(doc_id):
    url = f"https://acsearch.acr.org/docs/{doc_id}/Narrative/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
    }
    
    try:
        response = requests.get(url, headers=headers)
        
        if "application/pdf" in response.headers.get("Content-Type", ""):
            filename = f"acr_pdfs/doc_{doc_id}.pdf"
            with open(filename, "wb") as f:
                f.write(response.content)
            print(f"Downloaded {url} to {filename}")
            return True
        else:
            print(f"URL {url} did not return a PDF")
            return False
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return False

for doc_id in known_ids:
    success = download_pdf(doc_id)
    time.sleep(2)
```
